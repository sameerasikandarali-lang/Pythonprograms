###**Aim:To check that all braces are balanced or not**

[![Open In Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/sameerasikandarali-lang/pythonprograms/blob/main/codelinting.ipynb)

```python
def check_braces(code):
    stack = []
    pairs = {'}': '{', ')': '(', ']': '['}

    for i, char in enumerate(code):
        if char in "{[(":
            stack.append((char, i))

        elif char in ")]}":
            if not stack:
                print(f"Error: Unmatched {char} at position {i}")
                return False

            top, pos = stack.pop()
            if pairs[char] != top:
                print(f"Error: Mismatch at position {i}")
                return False

    if stack:
        for char, pos in stack:
            print(f"Error: Unmatched {char} at position {pos}")
        return False

    print("All braces are balanced")
    return True


code = input("Enter code: ")
check_braces(code)
```
