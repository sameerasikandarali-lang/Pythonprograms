#**Boolean operator**

```python
x = True
y = False

print(x and y)   # False
print(x or y)    # True
print(not x)     # False
```

#**Conversion/type casting**

```python
num_str = "100"
num_int = int(num_str)   # string to integer
num_float = float(num_str)  # string to float

print(num_int + 50)    # 150
print(num_float + 0.5) # 100.5
```
