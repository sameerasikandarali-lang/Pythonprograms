```python
import dill as dl #calling the module
var1=11.55
var2=3.45
var3=var1+var2 #create the variable var3
dl.dump_session('three_vars.pkl')# saves allthree variables
print(var3)
del var1, var2,var3 #deleting all the variables from python session
#print(var3)#'variable is no more'
```

```python
dl.load(open('three_vars.pkl','rb')) #load the pikle for consumption
print(var3)#now var3 isready to serve
```
