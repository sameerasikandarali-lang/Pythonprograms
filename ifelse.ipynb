#**Even or Odd**

```python
# Take number input from user
num = int(input("Enter a number: "))

# Check if number is divisible by 2
if num % 2 == 0:
    print("Even number")   # Runs when remainder is 0
else:
    print("Odd number")    # Runs when remainder is not 0
```

#**Positive or neggative**

```python
# Take number input
num = int(input("Enter a number: "))

# Check if number is greater than or equal to 0
if num >= 0:
    print("Positive number")   # Runs when number is positive or zero
else:
    print("Negative number")   # Runs when number is less than 0
```

#**Age check**

```python
# Take age input
age = int(input("Enter your age: "))

# Check voting eligibility
if age >= 18:
    print("You can vote")      # Runs if age is 18 or more
else:
    print("You cannot vote")   # Runs if age is less than 18
```
