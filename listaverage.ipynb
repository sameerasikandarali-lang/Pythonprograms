#**List and conditions(fruits)**

```python
# Creating a list of fruits
fruits = ['apple', 'banana', 'orange']

# Loop through the list
for i in fruits:
    print(i)   # print each fruit

    # Check condition
    if i == "orange":
        print("Current food value is:", i)   # print when orange found
```

#**Average of list**

```python
# List of numbers
Num = [3, 4, 2]

sum = 0   # variable to store total

# Loop to add numbers
for i in Num:
    print(i)        # print each number
    sum = sum + i   # add number to sum

# Calculate average
avg = sum / len(Num)

print("Average =", avg)   # display result
```
