```python
# Base class: Bird
class Bird:
    def __init__(self, species, color):
        self.species = species
        self.color = color

    def make_sound(self):
        print(f"{self.species} makes a sound.")

    def fly(self):
        print(f"{self.species} is flying!")


# Derived class: Sparrow
class Sparrow(Bird):
    def make_sound(self):
        print("Sparrow says: Chirp Chirp!")

# Derived class: Penguin
class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly!")

# Create instances
bird = Bird("Generic Bird", "Gray")
sparrow = Sparrow("Sparrow", "Brown")
penguin = Penguin("Penguin", "Black & White")

# Call methods
bird.make_sound()
bird.fly()

sparrow.make_sound()
sparrow.fly()

penguin.make_sound()
penguin.fly()
```
