#**for loops in python**

```python
# Print numbers from 1 to 10

for i in range(1, 11):
    print(i)
```

```python
#print number with message
for i in range(1,6):
  print("number is:", i)  #prints message + number
```

#**while loop with message**

```python
#while loop
i= 1
while i <= 5:
  print("count =", i)  #message + number
  i= i+1
```

#**print letters of a word**

```python
# Print each letter of a word

name = "Sameera"   # storing string

for ch in name:    # loop through each character
    print(ch)      # print character
```
