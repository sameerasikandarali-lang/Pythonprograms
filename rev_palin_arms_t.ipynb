#**Reverse number**

```python
# Function to reverse a number
def reverse_number(num):
    # Convert the number to string, reverse it, then convert back to integer
    return int(str(num)[::-1])

# Original number
num = 123

# Call the function and store the result
reversed_num = reverse_number(num)

# Display the reversed number
print("Reverse number:", reversed_num)
```

#**Reverse number taking user input**

```python
# Take number input from the user
num = int(input("Enter a number: "))

# Variable to store the reversed number
reverse = 0

# Loop runs until the number becomes 0
while num > 0:
    digit = num % 10            # Get the last digit of the number
    reverse = reverse * 10 + digit  # Add digit to reversed number
    num = num // 10             # Remove the last digit from number

# Print the reversed number
print("Reverse number:", reverse)
```

#**Palindrome**

```python
# Take input from user
num = int(input("Enter a number: "))

# Store original number
original = num

# Variable to store reversed number
reverse = 0

# Reverse the number
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

# Check palindrome
if original == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome")
```

#**Armstrong number**

```python
# Take input from user
num = int(input("Enter a number: "))

# Store original number
original = num

# Variable to store sum of powers
sum = 0

# Find number of digits
n = len(str(num))

# Calculate sum of digits raised to power n
while num > 0:
    digit = num % 10
    sum = sum + digit ** n
    num = num // 10

# Check Armstrong condition
if sum == original:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```
